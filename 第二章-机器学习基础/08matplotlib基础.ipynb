## matplotlib基础

```python
# 完全引入matplotlib
import matplotlib as mpl
# 在此课程中，仅仅使用到了它的一个子集
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

### 01 绘制曲线图

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
# 绘图，传入两个参数，第一个参数为横坐标值，第二个参数为纵坐标值
plt.plot(x, y)
```

#### 绘制多条曲线

```python
cosy = np.cos(x)
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
```

#### 可以传入颜色参数

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
```

#### 可以改变线条样式

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
```

#### 可以指定y轴和x轴的范围

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.xlim(0, 6.28)
```

#### 也可以使用axis函数一次性指定x轴和y轴的范围

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.axis([-1,11, -2,2])
```

#### 给图表添加字符串，使用xlabel，ylabel

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.xlabel("x axis")
plt.ylabel("y values")
```

#### 在函数图中标注曲线

```python
# 使用label属性
plt.plot(x, siny, label="sinx")
plt.plot(x, cosy, label="cosx")
# 注意！！！ 要加上这一句函数才会显示
plt.legend()
```

#### 给函数图取标题

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.title("Welcome to machine learning world!")
```
